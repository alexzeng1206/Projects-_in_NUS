## 1.0  Reading Text File

The basic way to access a file is to "open" it and assign it to a file object.  Here "r" specifies that it is a read file.  You cannot write to it - which is just as well if it is your source data.

The first statement just puts contents of file into the object called stockFile.

To read it, you apply the .read() method

Note that the output shows all the formatting "\"

```python
stockFile = open("STI_Index.txt","r")
stockFile.read()
```

To actually have a proper output, you can use the print function.  The print function will use the format in the appropriate manner.

Note that you must open and assign the contents of the file to stockFile once again since stockFile is actually empty after the last .read() command

```python
stockFile = open("STI_Index.txt","r")
print(stockFile.read())
```

You may also read the file line by line using the .readline() method

The data in the format as read in is very difficult to process.  For example, how do you total up the value of all the 30 stocks?
We will convert the data into a format such that each piece can be accessed easily.
We will use the .split() method.  We split according to separator ('\n') between each line of data.
You will note that a list results from using the .split method

```python
stockFile = open("STI_Index.txt","r")
stockList = stockFile.read().split('\n')
type(stockList)
```

You will notice that the formatting codes '\n' is no longer in the list.  However, the list contains 32 items (including header and a blank line at the end).  Each row contains data for one stock in string format.  It is easier to access individual row of data is you know the index as illustrated below.

```python
len(stockList)
```

```python
stockList
```

```python
stockList[1]  # data for OCBC bank
```

```python
print(stockList[1][0])  #first item in for the OCBC data
print(stockList[1][1])
```

You will note that within each row, it is still difficult to access the specific piece of data we need.  Each row is a single string.  But note that each piece of data within each row (other than the header row which we will ignore) is separated by the tab format code ('\t').  We can make use of this format code to separate the single strings in each rows to a list that we can access. 


```python
stockListBody = stockList[1:31] #just create a list that we want to work with - without header and the last empty row
print(len(stockListBody))
stockListBody
```

```python
stockListBody = stockList[1:31]
for i in range(len(stockListBody)):
    stockListBody[i]=stockListBody[i].split('\t')    #take each row and split the string up according to the '\t' formatting code 
stockListBody
```

```python
len(stockListBody)  #list of 30 stocks
```

```python
len(stockListBody[0])  # number of data items for each stock.
```

```python
stockListBody[0]  # accessing the first row of the list.  This is the data for the OCBC bank
```

```python
stockListBody[0][12]  # accessing the 13th item which is the value of the stock for the first list (which is OCBC bank)
```

With this transformation of the original data into a list of lists, we can now access the data and do some analysis.  For example, we can compute the sum of the total market values of all the 30 stocks. 

```python
totalValue = 0                       #initialize the variable we are using to sum up the market value of individual stocks
for i in range(len(stockListBody)):  # cycle through the entire list of 30 stocks
    totalValue = totalValue + float(stockListBody[i][12].replace(',',''))  # add value of specific stock to existing total value.  get rid of ',' in the string, change string to a float (numeric)
    print('{:10} {:20} {:20} {:20}'.format('Stock: ', stockListBody[i][0], 'Market Value:',stockListBody[i][12]))  #print out in orderly formatted manner
print('\n')   #print a blank line
print ('{:50} {} {:0,.0f}'.format('Total Value of 30 STI Stocks: ', '$', totalValue))  #print the total makret value, inserting the correct format. format for totalValue is {:0,.0f} - first 0 means no padding, start from the left, .0f means print as number with commas for thousands and with 0 decimal places
```

## 2.0  Writing A Text File

```python
with open("MarketValue.txt", "w") as file:    #using the with method to open a file results in the file closing autmoatically after we are done
    totalValue = 0                       
    for i in range(len(stockListBody)):  
        totalValue = totalValue + float(stockListBody[i][12].replace(',',''))  
        file.write( 'Stock: '+ '\t'+ stockListBody[i][0] + '\t'+ 'Market Value:'+ '\t' + stockListBody[i][12]+ '\n')  #writing all the output as one 
    file.write('Total Value of 30 STI Stocks: '+'\t'+ '$'+ str(totalValue))
```

Another way is to assign the file object using the normal assingment method (=).  However you must remember to write the code to close the file when you are done.

```python
file = open("MarketValueAlt.txt", "w")  #"w" means to write - will overwrite if existing file.  If you wish to append, use "a"
totalValue = 0
file.write('{:20}{:20}'.format('Stock','Market Value')+'\n')
for i in range(len(stockListBody)):  
    totalValue = totalValue + float(stockListBody[i][12].replace(',',''))  
    file.write('{:20} {:20}'.format(stockListBody[i][0],stockListBody[i][12]+ '\n'))
file.write('{:20} {} {:0,.0f}'.format('Total','$', totalValue))
file.close()
```

A simple way of writing to the file after you have tested the output in the console is just to direct the output to a file.  
After directing, the print out will send it to the file that you have re-directed it to.

Make sure you re-direct the output back to the console when done.

```python
import sys
file=open("MarketValueReroute.txt", "w")
old = sys.stdout    #save the old value of the sys.stdout so that it can be reset back later
sys.stdout = file   #redirecting the output to a file 
totalValue = 0                       
for i in range(len(stockListBody)):  
    totalValue = totalValue + float(stockListBody[i][12].replace(',',''))  
    print('{:10} {:20} {:20} {:20}'.format('Stock: ', stockListBody[i][0], 'Market Value:',stockListBody[i][12]))  #print out in orderly formatted manner
print('\n')   #print a blank line
print ('{:50} {} {:0,.0f}'.format('Total Value of 30 STI Stocks: ', '$', totalValue))  
file.close()
sys.stdout = old   #redirecting output back to the old output which is the console
```

## 3.0  Reading Text File as Table

It is much easier to read the text file, which is tab-delimited, as a table.  Saves all the trouble for pre-processing.

```python
import pandas as pd
stock = pd.read_table('STI_Index.txt')
```

We receive an error message that basically says that the implicit default assumption of the file being encoded in 'utf-8' format is untrue.  So we need to find out the encoding of the file so that we can use the right decoder.

The library chardet allows us to determine the encoding of a file.
The use is illustrated below.

```python
import chardet
fh = open("STI_Index.txt", "rb").read()
chardet.detect(fh)
```

After finding out the encoding scheme of the file, we use the encoding parameter in the read_table method to read the file.

```python
import pandas as pd
stock = pd.read_table('STI_Index.txt', encoding = 'ISO-8859-1')
stock
```
